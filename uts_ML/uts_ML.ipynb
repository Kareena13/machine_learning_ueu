```python
import pandas as pd
import matplotlib.pyplot as plt

product_a = ['teh pucuk', 'beng beng', 'choki-choki', 'kopiko', 'torabika', 'energen']
stok_counts = [52, 39, 78, 13, 78, 29]

df_names = pd.DataFrame(product_a, columns =['names'])
df_stok = pd.DataFrame(stok_counts, columns = ['stok'])

df = df_names.assign(stok = df_stok)

print(df)
```

```python
ax = plt.gca()
df.plot(kind='line',x='names',y='stok', color='blue', ax=ax, marker='o')

plt.show()
```

```python
import matplotlib.pyplot as plt

flavors = ('Teh pucuk','Beng beng','Choki-Choki', 'Kopiko','Torabika','energen')
votes = (78, 78, 13, 52, 39,29)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0, 0,  0.1 ,0, 0,0)

plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
)
plt.show()
```
